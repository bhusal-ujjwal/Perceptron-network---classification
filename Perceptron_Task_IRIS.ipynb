```python
import numpy as np

class MyPerceptron:
  #here we create the constructor of the class object
  def __init__(self, learning_rate=0.01, num_epochs=1000):
    self.learning_rate = learning_rate
    self.num_epochs = num_epochs
    self.weights = None
    self.bias = None

  #this function uses for fitting the data to perceptron, basically the training or setting correct weights
  def fit(self, X, y):
    self.weights = np.zeros((X.shape[1],)) #shape[1] is number of features we need a zero for weight per input 
    self.bias = 0

    for epoch in range(self.num_epochs):
      for i in range(X.shape[0]): # shape[0] is all the samples
        activation = np.dot(X[i], self.weights) + self.bias
        if (y[i]*activation) <= 0:
          self.weights += self.learning_rate * y[i] * X[i]
          self.bias += self.learning_rate * y[i]

  def predict(self, X):
      #compute prediction for the input data (usually the test set)
      predictions = np.dot(X, self.weights) + self.bias


      #apply the step function to convert the prediction to binary labels ( -1 or 1)
      predictions [ predictions <= 0 ] = -1
      predictions [ predictions > 0 ]  = 1


      return predictions

```

```python
# make test and train set for do the test and training
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,:-1].to_numpy() # means all the rows for all columns except last one which is target
y = df.iloc[:,-1].to_numpy() # means all the rows only for the last coloumn
```

```python
X
```

```python
#here we split the X and y to four parts the train parts for training the test parts for testing the model later
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
#fitting the perceptron object we just wrote
#we will create an object out of it and then fit it
perceptron = MyPerceptron() # note that when you create MyPerceptron object you can set parameters of learning_rate=0.01, num_epochs=1000 based on your will
perceptron.fit(X_train,y_train)
```

```python
#now that i trained my perceptron I have wights where i can used for prediction of test part
my_pred = perceptron.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
#this library is used for comparing my prediction of my model with the true target variables
```

```python
accuracy = accuracy_score(y_test,my_pred) # comparing my prediction with true targets
```
